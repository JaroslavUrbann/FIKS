```python
from datetime import date
import time

class Convertor:
    def __init__(self):
        self.authoritarian_definition()
        self.calculate_q()

        
    def authoritarian_definition(self):
        # start of authoritarian calendar
        self.d = 20
        self.m = 8
        self.y = 1984
        
        self.leap_when_dividable_by = 3
        self.not_leap_when_dividable_by = 300
        self.year_calendar = [25, 21, 21, 24, 24, 25, 25, 21, 25, 24, 21, 24, 21, 24, 25]
        self.leap_year_calendar = [25, 21, 22, 24, 24, 25, 25, 21, 25, 24, 21, 24, 21, 24, 25]
        self.days_in_a_year = sum(self.year_calendar)

        # number of gregorian leap years up to the start of the authoritarian calendar
        self.leap_years = self.y // 4 - self.y // 100 + self.y // 400 - (14 - self.m) // 12
        
    
    def calculate_q(self):
        a = (14 - self.m) // 12
        m = self.m + 12 * a - 3
        self.q = self.d + (153 * m + 2) // 5 - a * 365 - a // 4 + a // 100 - a // 400


    def gregorian2authoritarian_day(self, d, m, y):
        a = (14 - m) // 12
        leap_years = y // 4 - y // 100 + y // 400 - a
        leap_years -= self.leap_years
        y = y - self.y - a
        m = m + 12 * a - 3
        return d + (153 * m + 2) // 5 + 365 * y + leap_years - self.q
    
    
    def authoritarian_day2authoritarian(self, authoritarian_day):
        a_day = authoritarian_day

        y = a_day // self.days_in_a_year
        a_day -= y // self.leap_when_dividable_by + y // self.not_leap_when_dividable_by
        
        y2 = a_day // self.days_in_a_year
        
        if y != y2:
            a_day = authoritarian_day - y2 // self.leap_when_dividable_by + y2 // self.not_leap_when_dividable_by

        year = a_day // self.days_in_a_year + 1
        day = a_day % self.days_in_a_year + 1

        calendar = self.year_calendar
        if year % self.leap_when_dividable_by == 0 and year % self.not_leap_when_dividable_by != 0:
            calendar = self.leap_year_calendar
        
        month = 1
        for m in calendar:
            if day - m < 1:
                break
            day -= m
            month += 1
        
        print(authoritarian_day % 9 + 1)
        print(day)
        print(month)
        print(year)


c = Convertor()
day = c.gregorian2authoritarian_day(29, 8, 1984)
c.authoritarian_day2authoritarian(day)

# TODO: test gregorian2authoritarian_day & authoritarian_day2authoritarian (if zeros at least)
# d = date(1984, 8, 20)
# d2 = date(4000, 5, 15)
# for i in range(1):
#     day = c.gregorian2authoritarian_day(29, 8, 1984)
#     print((d2 - d).days)
#     print(day)
```
