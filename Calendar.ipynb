```python
class Convertor:
    def __init__(self):
        self.authoritarian_definition()
        self.calculate_q()

        
    def authoritarian_definition(self):
        # start of authoritarian calendar
        self.d = 20
        self.m = 8
        self.y = 1984
        
        months = [25, 21, 21, 24, 24, 25, 25, 21, 25, 24, 21, 24, 21, 24, 25]
        self.max_month_length = max(months)
        # [0, 4, 8, 9, 10, 10, 10, 14, 14, 15, 19, 20, 24, 25, 25]
        self.month_increments = [0]
        
        for i in range(len(months)):
            x = self.max_month_length - months[i] + self.month_increments[i]
            self.month_increments.append(x)

        # number of gregorian leap years up to the start of the authoritarian calendar
        a = (14 - self.m) // 12
        self.leap_years = self.y // 4 - self.y // 100 + self.y // 400 - a
        
    
    def calculate_q(self):
        a = (14 - self.m) // 12
        m = self.m + 12 * a - 3
        self.q = self.d + (153 * m + 2) // 5 - a * 365 - a // 4 + a // 100 - a // 400 - 1


    def gregorian2authoritarian_day(self, d, m, y):
        a = (14 - m) // 12
        leap_years = y // 4 - y // 100 + y // 400 - a
        leap_years -= self.leap_years
        y = y - self.y - a
        m = m + 12 * a - 3
        return d + (153 * m + 2) // 5 + 365 * y + leap_years - self.q
    
    
    def authoritarian_day2gregorian(self, authoritarian_days):
        a_d = authoritarian_days - 1
        y = a_d // 365
        leap_years = y // 4 - y // 100 + y // 400
        a_d -= leap_years
        y2 = a_d // 365
        
        if y != y2:
            leap_years = y2 // 4 - y2 // 100 + y2 // 400
            a_d = authoritarian_days - 1 - leap_years
            y = a_d // 365
        
        d = a_d % 365
        print(d)
        print(y)
    
    
    def authoritarian_day2authoritarian(self, authoritarian_days):
        a_d = authoritarian_days - 1
        y = a_d // 350
        
        leap_years = y // 3 - y // 300
        a_d -= leap_years
        
        y2 = a_d // 350
        d = a_d % 350
        
        if y != y2:
            leap_years = y2 // 3 - y2 // 300
            a_d = authoritarian_days - leap_years - 1
    
        year = a_d // 350 + 1
        d = a_d % 350 + 1
        
        m = d // self.max_month_length + 1
        d += self.month_increments[m]
        month = d // self.max_month_length + 1
        day = d % self.max_month_length
        

        
        if year % 3 == 0 and y % 300 != 0 and month > 3:
            day += 1
        
        print(day)
        print(month)
        print(year)


c = Convertor()
day = c.gregorian2authoritarian_day(15, 5, 4000)
#c.authoritarian_day2gregorian(day)
c.authoritarian_day2authoritarian(day)
```
